<h2 dir=rtl>
بازآرایی
</h2>

<div dir=rtl>
<b>تعریف:</b>
    بازآرایی به معنای تغییر در ساختار داخلی نرم‌افزار است به طوری که بدون آن که رفتار بیرونی آن تغییر کند، درک آن ساده‌تر و تغییر آن کم هزینه‌تر شود.
</div>

<div dir=rtl>
<b>سوال:</b>
    چرا باید بازآرایی کنیم؟
</div>

<div dir=rtl>
<ul>
<li>بازآرایی طراحی نرم‌افزار را بهبود می‌بخشد.</li>
<li>بازآرایی سبب ساده‌تر شدن فهم نرم‌افزار می‌شود.</li>
<li>بازآرایی به پیدا کردن اشکال در برنامه کمک می‌کند.</li>
<li>بازآرایی سبب افزایش سرعت برنامه‌نویسی می‌شود!</li>
</ul>
</div>

<div dir=rtl>
<b>سوال:</b>
چه زمانی باید بازآرایی کنیم؟
</div>

<div dir=rtl>
<ul>
<li>بار سومی که یک کار مشابه را انجام می‌دهید (<code>The Rule of Three</code>).</li>
<li>هر زمان که یک قابلیت (مانند تابع) به برنامه اضافه می‌کنید.</li>
<li>هر زمان که نیاز بود اشکالی را بر طرف کنید.</li>
<li>هنگام بازبینی کد (<code>Code Review</code>) بازآرایی انجام دهید.</li>
</ul>
</div>

<h3 dir=rtl>
نشانه‌های کد بد
</h3>

<div dir=rtl>
کد متعفن
<code>(Bad Smell)</code>
علائم هشدارآمیزی‌اند که به کاربر اعلان می‌کنند که اکنون زمان بازآرایی است. 
توجه کنید که این علائم صرفاً هشدار است. ممکن است برنامه‌نویس عمداً اینگونه برنامه‌نویسی کرده باشد و نیازی به تغییر در کد نباشد.
</div>

<div dir=rtl>
در ادامه با ۲۰ نشانه کد بد آشنا می‌شویم:
</div>

<div dir=rtl>
1. <b>کد تکراری: (<code>Duplication Code</code>)</b>
</div>

<div dir=rtl>
کد تکراری به این معنی است که در بخش‌های نرم‌افزار کدهایی می‌بینیم که مشابه یکدیگر اند. برای حذف کد تکراری معمولاً باید یک تابع استخراج شود و بخش‌هایی که کد تکراری وجود دارد حذف شده و به جای آن تابع فراخوانی شود.
  مشکل کد تکراری آن است که تغییرات منتشر‌شونده وجود دارد. اگر بعداً تصمیم بگیریم که کد را تغییر دهیم، مجبور هستیم که تمام بخش‌های تکراری را نیز تغییر دهیم. به همین خاطر است که نشانه کد تکراری امّ‌الفساد است.
</div>

<div dir=rtl>
    2. <b>متد طولانی: (<code>Long Method</code>)</b>    
</div>

<div dir=rtl>
فهمیدن متد طولانی سخت است. در نتیجه تغییر آن نیز مشکل است. از طرف دیگر متد طولانی معمولاً یک کار انجام نمی‌دهد. بنابراین نیاز است که این متد به چندین متد ریزدانه شکسته شود.
</div>

<div dir=rtl>
3. <b>کلاس بزرگ (<code>Large Class</code>)</b>
</div>

<div dir=rtl>
فهمیدن کلاس‌های بزرگ همانند متدهای طولانی سخت است. وقتی در یک سیستم کلاس بزرگ دیدید، احتمال زیاد کلاس‌هایی در سیستم وجود دارد که رفتار ندارند و صرفاً شامل داده اند.
برنامه‌نویسان زمانی که نمی‌دانند یک متد متعلق به کدام کلاس باید باشد، آن را به کلاس بزرگ منتقل می‌کنند.
به همین خاطر است که در بسیاری از سیستم‌ها
<code>utility class</code>
دیده می‌شود. در صورتی که اگر به درستی شیء‌گرا کد بنویسید، نباید چنین کلاسی دیده شود.
قاعده کلی آن است که رفتار باید به سمت داده برود. به عبارت دیگر ببینید که داده متد در کدام کلاس قرار دارد و متد را به آن کلاس منتقل کنید. اگر دیدید که هیچ کلاسی داده مورد نظر را ندارد، احتمالاً به این معنا است که باید
یک کلاس جدید تعریف شود.
در واقع در شیء‌گرایی داده محور اصلی است.
</div>

<div dir=rtl>
4. <b>لیست پارامترهای طولانی: (<code>Long Parameter List</code>)</b>
</div>

<div dir=rtl>
یکی از اصول شیء‌گرایی
<code>SOLID</code>
اصل جداسازی واسطه‌ها است
(<code>Interface Segregation Principle</code>).
این اصل می‌گوید به جای آن یک متد داشته باشید که تعداد زیادی پارامتر دریفات می‌کند. چندین متد داشته باشید که هر کدام یک کار ریزدانه انجام می‌دهند و تعداد پارامترها کم است.
</div>

<div dir=rtl>
5. <b>تغییرات واگرا: (<code>Divergent Change</code>)</b>
</div>

<div dir=rtl>
زمانی که یک کلاس به دلايل بی‌ربط تغییر می‌کند، شاهد این نشانه کد بد هستیم. این نشانه دلالت بر نقض اصل
تک مسئولیتی
<code>(Single Responsibility Principle)</code>
است.
</div>

<div dir=rtl>
6. <b><code>Shotgun Surgery</code></b>
</div>

<div dir=rtl>
وقتی که هر زمان تغییری در یک بخش از سیستم بدهید، لازم شود که تعداد زیادی کلاس دیگر را تغییر دهید، به این معنا است که تغییرات منتشرشونده دارید.
درست مانند شلیک شاتگان که گلوله‌ها در جاهای مختلف بدن پخش می‌شود. این نشانه سبب می‌شود که قابلیت نگهداشت سیستم به خاطر وابستگی زیاد کلاس‌ها به یکدیگر به شدت پایین بیاید.
</div>

<div dir=rtl>
7. <b>حسادت ویژگی: (<code>Feature Envy</code>)</b>
</div>

<div dir=rtl>
دو کلاس
<code>A</code>
و
<code>B</code>
را تصور کنید. فرض کنید متد
<code>m</code>
داخل کلاس
<code>A</code>
قرار دارد.
ولیکن این متد به نظر می‌رسد که بیشتر با داده‌های کلاس
<code>B</code>
کار می‌کند تا با کلاس خودش. به عبارت دیگر متد به آن یکی کلاس بیشتر علاقمند است.
در این صورت ممکن است جای متد درست تعریف نشده باشد و متد
<code>m</code>
در واقع متعلق به کلاس
<code>B</code>
بوده باشد. همان طور که پیش‌تر گفتیم، در شء‌گرایی داده محوریت است. رفتار باید کنار داده‌اش قرار بگیرد.
</div>

<div dir=rtl>
8. <b>توده داده: (<code>Data Clumps</code>)</b>
</div>

<div dir=rtl>
زمانی که دسته‌ای از داده‌ها به صورت منظم با یکدیگر ظاهر می‌شوند.
در این حالت باید تبدیل به کلاس شوند. به عنوان نمونه فرض کنید تعداد زیادی متد داریم که پارامترهای طول و عرض را دریافت می‌کنند. هر جایی که پارامتر طول ظاهر شده است، پارامتر عرض هم دیده می‌شود.
در این صورت یک کلاس ایجاد کنید که شامل خصوصیت طول و عرض با یکدیگر باشد و از این به بعد کل کلاس را به عنوان پارامتر ارسال کنید.
</div>

<div dir=rtl>
9. <b><code>Primitive Obsession</code></b>
</div>

<div dir=rtl>
استفاده بیش از حد از داده‌های بسیط
مانند اعداد صحیح و اعشاری یکی دیگر از علائم هشدارآمیز است. به طور مثال فرض کنید که به جای استفاده از کلاسی مانند تاریخ، زمان‌ها را با اعداد صحیح نمایش دهیم.
در واقع نباید ترس از ایجاد کلاس داشته باشیم، هر چند که در ابتدا کلاس ریزدانه باشد. معمولاً بعد از مدتی کلاس‌ها رفتار پیدا می‌کنند و بزرگ می‌شوند. به عنوان نمونه تاریخ حتی تبدیل به یک کتابخانه شده است.
</div>

<div dir=rtl>
10. <b>تعداد زیادی <code>if-elif</code></b>
</div>

<div dir=rtl>
زمانی که تعداد زیادی عبارت شرطی در کد دیده می‌شود. پیچیدگی افزایش پیدا می‌کند. معمولاً چنین عبارت‌هایی بدین معنی است که از خواص شیء‌گرایی به درستی استفاده نشده است و چندریختی حذف شده است.
</div>

<div dir=rtl>
11. <b>ساختارهای توارثی موازی: (<code>Parallel Inheritance Hierarchies</code>)</b>
</div>

<div dir=rtl>
هر زمان که تغییری در یکی از زیرکلاس‌ها اعمال می‌کنید، مجبور شوید که در زیرکلاس روبرویی نیز تغییر ایجاد کنید، در این صورت ساختار توارثی موازی داریم که با یکدیگر رشد م‌کنند. به عبارت دیگر وابستگی بین دو ساختار وجود دارد و سبب تغییرات منتشر شونده می‌شود.
</div>

<div dir=rtl>
12. <b>کلاس تنبل: (<code>Lazy Class</code>)</b>
</div>

<div dir=rtl>
کلاسی که رفتاری از خود بروز نمی‌دهد و توجیه وجودی‌اش از دست رفته است. این کلاس معمولاً نتیجه اعمال مکرر بازآرایی بر روی کلاس بزرگ است و در نهایت باید حذف شود.
</div>

<div dir=rtl>
13. <b>عمومیت بر اساس حدس و گمان: (<code>Speculative Generality</code>)</b>
</div>

<div dir=rtl>
هر وقت قابلیتی به کد موجود اضافه کنیم به این امید که شاید در آینده یک روز نیاز پیدا کند، با این نشانه کد بد روبرو هستیم.
</div>

<div dir=rtl>
14. <b>خصوصیت موقتی: (<code>Temporary Field</code>)</b>
</div>

<div dir=rtl>
هر زمان خصوصیت تعریف شده تنها در زمان‌های خاصی مقداردهی می‌شود با این نشانه روبرو هستیم. در این صورت شاید نباید به صورت خصوصیت تعریف می‌شده است. بلکه 
صرفاً پارامتر ورودی یک تابع بوده است. حالت دیگر زمانی است که فقط بعضی از متدهای کلاس از خصوصیت استفاده می‌کنند و دیگران با آن کاری ندارند. در این صورت شاید به معنی آن باشد که باید
زیر کلاس تعریف کرد.
</div>

<div dir=rtl>
15. <b>زنجیره پیغام: (<code>Message Chains</code>)</b>
</div>

<div dir=rtl>
زمانی که زنجیره دید ترایا دیده شود، به این معنا است که رفتار از داده دور افتاده است. همچنین کپسوله‌سازی نقض شده است. راه حل این مسئله آن است که رفتار در طول زنجیره حرکت کند تا به داده خود برسد.
</div>

<div dir=rtl>
16. <b>صمیمت نامناسب: (<code>Inappropriate Intimacy</code>)</b>
</div>

<div dir=rtl>
زمانی که وابستگی بین کلاس‌ها بیش از حد باشد با این نشانه کد روبرو هستیم. مشکل این نشانه آن است که تغییرات منتشر شونده داریم.
</div>

<div dir=rtl>
17. <b>کلاس‌های جایگزین با واسط متفاوت</b>
</div>

<div dir=rtl>
بعضی از وقت‌ها در نتیجه تکامل سیستم با کلاس‌هایی روبرو می‌شویم که کارهای یکسان با واسط‌های متفاوت انجام می‌دهند. در عمل به نوعی کد تکراری دیده می‌شود.
</div>

<div dir=rtl>
18. <b>کلاس داده: (<code>Data Class</code>)</b>
</div>

<div dir=rtl>
کلاس‌هایی که تنها تعدادی خصوصیت به همراه 
<code>getter</code>
و
<code>setter</code>
دارند یکی از نشانه‌های کد بد است. معمولاً در این مواقع کلاس‌های بزرگ در سیستم دیده می‌شود. توسعه‌دهنده باید رفتارهای کلاس‌های بزرگ مرتبط با این نوع از کلاس‌ها را پیدا کرده و به این کلاس منتقل کند.
</div>

<div dir=rtl>
19. <b>میراث مردود: (<code>Refused Bequest</code>)</b>
</div>

<div dir=rtl>
زمانی که زیرکلاس تعهدات پدر را رعایت نمی‌کند.
به عبارت دیگر زمانی که زیرکلاس پیش‌شرط را قوی‌تر می‌کند یا پس‌شرط را ضعیف‌تر می‌کند دیگر نباید به عنوان زیرکلاس کلاس مورد نظر قرار بگیرد.
</div>

<div dir=rtl>
20. <b>کامنت</b>
</div>

<div dir=rtl>
کامنت ابزار بسیار خوبی برای توضیح چیستی قطعه کد نوشته شده است. ولیکن هرگز نباید کامنت به منظور جبران کد بد نوشته شده استفاده شود.
</div>

<h3 dir=rtl>
الگوهای بازآرایی: طبقه‌بندی
<h3>

<div dir=rtl>
در ادامه درس الگوهای بازآرایی را مورد واکاوی قرار می‌دهیم. این الگوها در پنج دسته طبقه‌بندی کرده‌ایم. این طبقه‌بندی‌ها عبارت‌اند از:
</div>

<div dir=rtl>
1. <b>ساختن متدها: (<code>Composing Methods</code>)</b>
    ناظر بر این است که چگونه رفتار را در قالب متدها دسته‌بندی کنیم.
</div>

<div dir=rtl>
2. <b>سازماندهی داده:</b>
    داده‌ها را چگونه سازماندهی کنیم تا راحت‌تر بتوان با آن کار کرد.
</div>

<div dir=rtl>
3. <b>ساده‌سازی عبارت‌های شرطی: (<code>Simplifying Conditional Expressions</code>)</b>
    عبارت‌های شرطی با هدف کاهش وقوع خطاهای احتمالی
</div>

<div dir=rtl>
4. <b>ساده‌سازی فراخوانی‌های متد</b>
</div>

<div dir=rtl>
5. <b>برخورد با ساختارهای توارثی:</b>
    ناظر بر انتقال خصوصیات و رفتار بین ساختارهای توارثی است.
</div>
